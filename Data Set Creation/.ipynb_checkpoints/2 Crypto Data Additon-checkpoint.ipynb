```python
from datetime import datetime
import pandas as pd
import os
from IPython.display import display
from pycoingecko import CoinGeckoAPI
```

# Pulling CoinGecko Data for Bitcoin Price

```python

cg = CoinGeckoAPI()

start_date = datetime(2023, 12, 1)
end_date = datetime(2024, 2, 29)

start_timestamp = int(start_date.timestamp())
end_timestamp = int(end_date.timestamp())

btc_price_data = cg.get_coin_market_chart_range_by_id(
    id='bitcoin',
    vs_currency='usd',
    from_timestamp=start_timestamp,
    to_timestamp=end_timestamp
)

btc_df = pd.DataFrame(btc_price_data['prices'], columns=['timestamp', 'price'])
btc_df['timestamp'] = pd.to_datetime(btc_df['timestamp'], unit='ms')

btc_daily_df = btc_df.set_index('timestamp').resample('D').mean().reset_index()

btc_daily_df.head()

```

```python
combined_df = pd.read_csv('/Users/jessiefreelander/Flatiron/Git_lab_Phase Project/Phase-4-Project/Phase 4 project Data/combined_data.csv')

combined_df
```

# Discover Data

```python
# Extract the prices from btc_price_data
prices = btc_price_data['prices']

# Convert to a DataFrame for easier analysis
prices_df = pd.DataFrame(prices, columns=['timestamp', 'price'])

# Convert the timestamp to datetime
prices_df['timestamp'] = pd.to_datetime(prices_df['timestamp'], unit='ms')

# Display the first few rows of the prices DataFrame
print(prices_df.head(20))

# Check the max and min prices
max_price = prices_df['price'].max()
min_price = prices_df['price'].min()

print(f"Maximum Price: {max_price}")
print(f"Minimum Price: {min_price}")
```

# Merging Tweet and Crypto Data Sets


```python

combined_df['date'] = pd.to_datetime(combined_df['date'], errors='coerce')

btc_daily_df['date'] = pd.to_datetime(btc_daily_df['timestamp'], unit='ms').dt.strftime('%m/%d/%Y')

combined_df['date'] = combined_df['date'].dt.strftime('%m/%d/%Y')
btc_daily_df['date'] = btc_daily_df['date'].astype(str)

merged_df = pd.merge(combined_df, btc_daily_df[['date', 'price']], on='date', how='inner')
merged_df
```

# Discovering 24 Hour Price Change


```python
btc_daily_df['24 Change'] = btc_daily_df['price'].pct_change(periods=1) * 100

btc_daily_df.head(10)
```

# Discovering Peak to Trough Volatility Percentage wihtin 24 Hour Periods

```python

prices_df['date'] = pd.to_datetime(prices_df['timestamp'].dt.date)

peak_trough_df = prices_df.groupby('date')['price'].agg(['max', 'min']).reset_index()

peak_trough_df['peak_to_trough_volatility'] = ((peak_trough_df['max'] - peak_trough_df['min']) / peak_trough_df['min']) * 100

print(peak_trough_df)
```

# Merging Data Sets

```python
merged_df['date'] = pd.to_datetime(merged_df['date'], errors='coerce')

btc_daily_df['date'] = pd.to_datetime(btc_daily_df['date'], errors='coerce')


merged_df = pd.merge(merged_df, btc_daily_df[['date', '24 Change']], on='date', how='left')

merged_df
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'], errors='coerce')

peak_trough_df['date'] = pd.to_datetime(peak_trough_df['date'], errors='coerce')

merged_df = pd.merge(merged_df, peak_trough_df[['date', 'peak_to_trough_volatility']], on='date', how='left')

merged_df.rename(columns={'peak_to_trough_volatility': 'Interday Volatility'}, inplace=True)

```

```python
merged_df.rename(columns={'price': 'open price'}, inplace=True)

merged_df
```

```python
output_file_path = '/Users/jessiefreelander/Flatiron/Git_lab_Phase Project/Phase-4-Project/Phase 4 project Data/merged_data.csv'
merged_df.to_csv(output_file_path, index=False)
```

# Plotting Bitcoin Price Over Time

```python

import matplotlib.pyplot as plt


btc_daily_df['date'] = pd.to_datetime(btc_daily_df['date'], format='%m/%d/%Y')


plt.figure(figsize=(12, 6))
plt.plot(btc_daily_df['date'], btc_daily_df['price'], label='Bitcoin Price', color='blue')
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
